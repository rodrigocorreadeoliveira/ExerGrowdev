#Exercícios

Qual a quantidade de pessoas nas seguintes faixas etárias
- 20 anos ou menos
- 21 a 35 anos
- 36 a 50 anos
- 51 a 65 anos
- maiores de 65 anos



```python
import pandas as pd

df = pd.read_csv('dados.csv')

display(df)

# 20 anos ou menos
menos_20 = df[df['Idade'] <= 20]['Idade'].count()

# 21 a 35 anos
vinteum_trintaecinco = df[(df['Idade'] > 20) & (df['Idade'] <= 35)]['Idade'].count()

# 36 a 50 anos
trintaeseis_cinquenta = df[(df['Idade'] > 35) & (df['Idade'] <= 50)]['Idade'].count()

# 51 a 65 anos
cinquentaeum_sessentaecinco = df[(df['Idade'] > 50) & (df['Idade'] <= 65)]['Idade'].count()

# maiores de 65 anos
mais_sessentaecinco = df[df['Idade'] > 65]['Idade'].count()

print("Pessoas com 20 anos ou menos:", menos_20)
print("Pessoas entre 21 e 35 anos:", vinteum_trintaecinco)
print("Pessoas entre 36 e 50 anos:", trintaeseis_cinquenta)
print("Pessoas entre 51 e 65 anos:", cinquentaeum_sessentaecinco)
print("Pessoas com mais de 65 anos:", mais_sessentaecinco)
```

Qual é a média de escolaridade em cada faixa etária?


```python

# 20 anos ou menos
media_escolaridade_menos_20 = df[df['Idade'] <= 20]['Anos de Estudo'].mean()

# 21 a 35 anos
media_escolaridade_vinteum_trintaecinco = df[(df['Idade'] > 20) & (df['Idade'] <= 35)]['Anos de Estudo'].mean()

# 36 a 50 anos
media_escolaridade_trintaeseis_cinquenta = df[(df['Idade'] > 35) & (df['Idade'] <= 50)]['Anos de Estudo'].mean()

# 51 a 65 anos
media_escolaridade_cinquentaeum_sessentaecinco = df[(df['Idade'] > 50) & (df['Idade'] <= 65)]['Anos de Estudo'].mean()

# maiores de 65 anos
media_escolaridade_mais_sessentaecinco = df[df['Idade'] > 65]['Anos de Estudo'].mean()

print("Média de escolaridade para pessoas com 20 anos ou menos: {:.2f} anos".format(media_escolaridade_menos_20))
print("Média de escolaridade para pessoas entre 21 e 35 anos: {:.2f} anos".format(media_escolaridade_vinteum_trintaecinco))
print("Média de escolaridade para pessoas entre 36 e 50 anos: {:.2f} anos".format(media_escolaridade_trintaeseis_cinquenta))
print("Média de escolaridade para pessoas entre 51 e 65 anos: {:.2f} anos".format(media_escolaridade_cinquentaeum_sessentaecinco))
print("Média de escolaridade para pessoas com mais de 65 anos: {:.2f} anos".format(media_escolaridade_mais_sessentaecinco))
```

Qual a UF que concentra a maior renda?


```python
# Tratamento dos dados de Sexo
df['Sexo'] = df['Sexo'].replace({0: 'Masculino', 1: 'Feminino'})

# Tratamento dos dados de Cor
df['Cor'] = df['Cor'].replace({0: 'Indígena', 2: 'Branca', 4: 'Preta', 6: 'Amarela', 8: 'Parda', 9: 'Sem declaração'})

# Tratamento dos dados de UF
df['UF'] = df['UF'].replace({ 11: 'Rondônia',
    12: 'Acre',
    13: 'Amazonas',
    14: 'Roraima',
    15: 'Pará',
    16: 'Amapá',
    17: 'Tocantins',
    21: 'Maranhão',
    22: 'Piauí',
    23: 'Ceará',
    24: 'Rio Grande do Norte',
    25: 'Paraíba',
    26: 'Pernambuco',
    27: 'Alagoas',
    28: 'Sergipe',
    29: 'Bahia',
    31: 'Minas Gerais',
    32: 'Espírito Santo',
    33: 'Rio de Janeiro',
    35: 'São Paulo',
    41: 'Paraná',
    42: 'Santa Catarina',
    43: 'Rio Grande do Sul',
    50: 'Mato Grosso do Sul',
    51: 'Mato Grosso',
    52: 'Goiás',
    53: 'Distrito Federal'})

display(df)
```

```python
# Agrupa os dados por UF e calcula a renda média para cada UF
renda_por_uf = df.groupby('UF')['Renda'].mean()

# Encontra a UF com a maior renda média
uf_maior_renda = renda_por_uf.idxmax()

print("A UF com a maior renda média é:", uf_maior_renda)

```

Qual a renda média de pessoas pretas e brancas de homens e mulheres?


```python
# Renda média de homens pretos
renda_media_homens_pretos = df[(df['Sexo'] == 'Masculino') & (df['Cor'] == 'Preta')]['Renda'].mean()

# Renda média de homens brancos
renda_media_homens_brancos = df[(df['Sexo'] == 'Masculino') & (df['Cor'] == 'Branca')]['Renda'].mean()

# Renda média de mulheres pretas
renda_media_mulheres_pretas = df[(df['Sexo'] == 'Feminino') & (df['Cor'] == 'Preta')]['Renda'].mean()

# Renda média de mulheres brancas
renda_media_mulheres_brancas = df[(df['Sexo'] == 'Feminino') & (df['Cor'] == 'Branca')]['Renda'].mean()

print("Renda média de homens pretos: {:.2f}".format(renda_media_homens_pretos))
print("Renda média de homens brancos: {:.2f}".format(renda_media_homens_brancos))
print("Renda média de mulheres pretas: {:.2f}".format(renda_media_mulheres_pretas))
print("Renda média de mulheres brancas: {:.2f}".format(renda_media_mulheres_brancas))
```

Qual a renda média concentrada em cada faixa etária?




```python
# 20 anos ou menos
renda_media_menos_20 = df[df['Idade'] <= 20]['Renda'].mean()

# 21 a 35 anos
renda_media_vinteum_trintaecinco = df[(df['Idade'] > 20) & (df['Idade'] <= 35)]['Renda'].mean()

# 36 a 50 anos
renda_media_trintaeseis_cinquenta = df[(df['Idade'] > 35) & (df['Idade'] <= 50)]['Renda'].mean()

# 51 a 65 anos
renda_media_cinquentaeum_sessentaecinco = df[(df['Idade'] > 50) & (df['Idade'] <= 65)]['Renda'].mean()

# maiores de 65 anos
renda_media_mais_sessentaecinco = df[df['Idade'] > 65]['Renda'].mean()

print("Renda média para pessoas com 20 anos ou menos: {:.2f}".format(renda_media_menos_20))
print("Renda média para pessoas entre 21 e 35 anos: {:.2f}".format(renda_media_vinteum_trintaecinco))
print("Renda média para pessoas entre 36 e 50 anos: {:.2f}".format(renda_media_trintaeseis_cinquenta))
print("Renda média para pessoas entre 51 e 65 anos: {:.2f}".format(renda_media_cinquentaeum_sessentaecinco))
print("Renda média para pessoas com mais de 65 anos: {:.2f}".format(renda_media_mais_sessentaecinco))
```

Qual a renda média nas faixas de anos de estudo abaixo?
- 2 anos ou menos
- 3 a 6 anos
- 7 a 10 anos
- 10 a 12 anos
- 13 anos ou mais


```python
# 2 anos ou menos
renda_media_estudo_menos_2 = df[df['Anos de Estudo'] <= 2]['Renda'].mean()

# 3 a 6 anos
renda_media_estudo_3_6 = df[(df['Anos de Estudo'] > 2) & (df['Anos de Estudo'] <= 6)]['Renda'].mean()

# 7 a 10 anos
renda_media_estudo_7_10 = df[(df['Anos de Estudo'] > 6) & (df['Anos de Estudo'] <= 10)]['Renda'].mean()

# 10 a 12 anos
renda_media_estudo_10_12 = df[(df['Anos de Estudo'] > 10) & (df['Anos de Estudo'] <= 12)]['Renda'].mean()

# 13 anos ou mais
renda_media_estudo_mais_13 = df[df['Anos de Estudo'] > 12]['Renda'].mean()

print("Renda média para pessoas com 2 anos ou menos de estudo: {:.2f}".format(renda_media_estudo_menos_2))
print("Renda média para pessoas com 3 a 6 anos de estudo: {:.2f}".format(renda_media_estudo_3_6))
print("Renda média para pessoas com 7 a 10 anos de estudo: {:.2f}".format(renda_media_estudo_7_10))
print("Renda média para pessoas com 10 a 12 anos de estudo: {:.2f}".format(renda_media_estudo_10_12))
print("Renda média para pessoas com 13 anos ou mais de estudo: {:.2f}".format(renda_media_estudo_mais_13))
```

Qual a escolaridade entre indígenas, para homens e mulheres?


```python
# Escolaridade média de homens indígenas
escolaridade_media_homens_indigenas = df[(df['Sexo'] == 'Masculino') & (df['Cor'] == 'Indígena')]['Anos de Estudo'].mean()

# Escolaridade média de mulheres indígenas
escolaridade_media_mulheres_indigenas = df[(df['Sexo'] == 'Feminino') & (df['Cor'] == 'Indígena')]['Anos de Estudo'].mean()

print("Escolaridade média de homens indígenas: {:.2f}".format(escolaridade_media_homens_indigenas))
print("Escolaridade média de mulheres indígenas: {:.2f}".format(escolaridade_media_mulheres_indigenas))
```

Qual a média de anos de estudo entre brancos e negros?


```python
# Média de anos de estudo de pessoas brancas
media_estudo_brancos = df[df['Cor'] == 'Branca']['Anos de Estudo'].mean()

# Média de anos de estudo de pessoas negras
media_estudo_negros = df[df['Cor'] == 'Preta']['Anos de Estudo'].mean()

print("Média de anos de estudo de pessoas brancas: {:.2f}".format(media_estudo_brancos))
print("Média de anos de estudo de pessoas negras: {:.2f}".format(media_estudo_negros))
```

Qual a média de anos de estudo e de renda para mulheres brancas, e para mulheres negras?

```python
# Mulheres brancas
media_estudo_mulheres_brancas = df[(df['Sexo'] == 'Feminino') & (df['Cor'] == 'Branca')]['Anos de Estudo'].mean()
media_renda_mulheres_brancas = df[(df['Sexo'] == 'Feminino') & (df['Cor'] == 'Branca')]['Renda'].mean()

# Mulheres negras
media_estudo_mulheres_negras = df[(df['Sexo'] == 'Feminino') & (df['Cor'] == 'Preta')]['Anos de Estudo'].mean()
media_renda_mulheres_negras = df[(df['Sexo'] == 'Feminino') & (df['Cor'] == 'Preta')]['Renda'].mean()

print("Média de anos de estudo de mulheres brancas: {:.2f}".format(media_estudo_mulheres_brancas))
print("Média de renda de mulheres brancas: {:.2f}".format(media_renda_mulheres_brancas))
print("Média de anos de estudo de mulheres negras: {:.2f}".format(media_estudo_mulheres_negras))
print("Média de renda de mulheres negras: {:.2f}".format(media_renda_mulheres_negras))
```

 Qual é a UF com maior média de escolaridade e qual a UF com maior média de renda?

```python
# UF com maior média de escolaridade
escolaridade_por_uf = df.groupby('UF')['Anos de Estudo'].mean()
uf_maior_escolaridade = escolaridade_por_uf.idxmax()

# UF com maior média de renda
renda_por_uf = df.groupby('UF')['Renda'].mean()
uf_maior_renda = renda_por_uf.idxmax()

print("A UF com maior média de escolaridade é:", uf_maior_escolaridade)
print("A UF com maior média de renda é:", uf_maior_renda)
```
